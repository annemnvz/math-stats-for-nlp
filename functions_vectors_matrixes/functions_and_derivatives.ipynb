4. Given the function $\mbox{ReLU}(x) = \left\lbrace
\begin{array}{cc}
x  \, , & \; x \ge 0 \\
0  \, , & \; x < 0 \\
\end{array}
 \right.$:
 - Draw it.
 - What do you think looks its derivative like?

```python
def ReLu(x):
    return(x>=0)*x

x = np.linspace(-3, 5, num=100)
y = ReLu(x)
plt.plot(x, y, '-')
#slope constant line at 0 and then increases linearly at f(x)=1
```
